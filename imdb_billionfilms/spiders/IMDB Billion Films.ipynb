```python
import pandas as pd
import numpy as np
import json
import matplotlib as mpl
import seaborn as sns
import re
import folium
import folium.plugins
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option('max_rows',999)
pd.set_option('float_format','{:,.0f}'.format)
```

```python
with open('movies.json','r') as f:
    output = json.loads(f.read())
```

```python
types = []
for y in output:
    types.append(y['type'])
types = np.unique(types)
```

```python
def to_dictofdataframes(output, types):
    dict_out = {}
    for t in types:
        i = 0;
        for y in output:
            if y['type'] == t:
                df_t = pd.Series(y).to_frame().T
                if i == 0:
                    df = df_t
                else:
                    df = df.append(df_t, ignore_index=True)
                i += 1
        dict_out[t] = df
    return dict_out
```

```python
dict_df = to_dictofdataframes(output, types)
```

```python
actors['bio'] = actors.bio.str.strip()
actors['height_m'] = actors.height.str.extract('(\d*\.\d*)', expand=False).astype(float)
actors['height_f'] = actors.height.str.extract('(\d)\'', expand=False).astype(int)
actors['height_in'] = actors.height.str.extract('(\d+)\D?\"', expand=False).fillna(0).astype(int)
actors['birthdate'] = pd.to_datetime(actors.birthdate)

def get_point(x):
    geocoder = Nominatim()
    try: point = geocoder.geocode(x).point
    except: return None
    return [point.latitude,point.longitude]

actors['birthplace_pt'] = actors.birthplace.apply(get_point)
actors['birth_lat'] = actors.birthplace_pt.apply(lambda x: x[0] if type(x)==list else None)
actors['birth_lng'] = actors.birthplace_pt.apply(lambda x: x[1] if type(x)==list else None)
actors['numquotes'] = actors.quotes.apply(lambda x: len(x))
actors['age'] = pd.to_datetime('today').year - actors.birthdate.dt.year
actors['numquotes_bin'] = pd.cut(actors.numquotes, 5)
```

# People Table

```python
actors_list = []
directors = []
writers = []

titles.actors.apply(lambda x: [actors_list.append(p) for p in x])
titles.directors.apply(lambda x: [directors.append(p) for p in x])
titles.writers.apply(lambda x: [writers.append(p) for p in x])

actors_list = list(np.unique(actors_list))
directors = list(np.unique(directors))
writers = list(np.unique(writers))
people = np.unique(actors_list+directors+writers)
```

```python
def get_roles(p):
    roles = []
    if p in actors_list:
        roles.append('actor')
    if p in directors:
        roles.append('director')
    if p in writers:
        roles.append('writer')
    return roles
```

```python
dict_people = {
    p: {
        'titles': [t for t, a, d, w in zip(titles.title,titles.actors,titles.directors,titles.writers) if p in a+d+w],
        'roles': get_roles(p)
    }
    for p in people
}
```

```python
people_df = pd.DataFrame(dict_people).T
people_df['numroles'] = people_df.roles.apply(lambda x: len(x))
people_df['numtitles'] = people_df.titles.apply(lambda x: len(x))
```

```python
actors = actors.merge(people_df, left_on='actor', right_index=True)
```

```python
actors = actors[[x for x in actors.columns if x not in ('images','height','birthplace_pt')]]
```
